# Telco Customer Churn — Dashboard Preparation Notebook

This notebook prepares the data needed for the Power BI dashboard.
It loads model predictions, enriches them with customer attributes,
and exports a clean dataset for visualization.

Outputs include:
- Churn probability scores
- Actual churn labels
- Key customer attributes
- Recommended dashboard structure

```python
import pandas as pd

df_clean = pd.read_csv("../data/cleaned/cleaned_telco_churn.csv")
preds = pd.read_csv("../data/modeling/churn_predictions_for_dashboard.csv")

dashboard_df = preds.merge(df_clean, on='customerID', how='left')
```

```python
# Adjust Churn Labels
dashboard_df['ChurnLabel'] = dashboard_df['ActualChurn'].map({0:'No', 1:'Yes'})
```

```python
# Risk Segmentation
dashboard_df['RiskSegment'] = pd.cut(
    dashboard_df['PredictedProb'],
    bins=[0, 0.25, 0.50, 0.75, 1.0],
    labels=['Low', 'Medium', 'High', 'Critical']
)
```

```python
# Selecting Dashboard Columns
dashboard_export = dashboard_df[[
    'customerID',
    'ActualChurn',
    'ChurnLabel',
    'PredictedProb',
    'RiskSegment',
    'tenure_group',
    'tenure_y',
    'MonthlyCharges_y',
    'TotalCharges_y',
    
    # Encoded categorical features
    'Contract_Month-to-month',
    'Contract_One year',
    'Contract_Two year',
    
    'PaymentMethod_Bank transfer (automatic)',
    'PaymentMethod_Credit card (automatic)',
    'PaymentMethod_Electronic check',
    'PaymentMethod_Mailed check',
    
    'InternetService_DSL',
    'InternetService_Fiber optic',
    'InternetService_No',
    
    'gender_Female',
    'gender_Male',
    
    'SeniorCitizen_y',
    'Partner_y',
    'Dependents_y'
]]
```

```python
#Rename Columns
dashboard_export = dashboard_export.rename(columns={
    'tenure_y': 'tenure',
    'MonthlyCharges_y': 'MonthlyCharges',
    'TotalCharges_y': 'TotalCharges',
    'SeniorCitizen_y': 'SeniorCitizen',
    'Partner_y': 'Partner',
    'Dependents_y': 'Dependents'
})
```

```python
#Export
dashboard_export.to_csv("../data/dashboard/dashboard_telco_churn.csv", index=False)
```
