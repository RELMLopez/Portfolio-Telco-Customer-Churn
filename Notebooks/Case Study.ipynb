# Telco Customer Churn — Case Study Notebook

This notebook summarizes the end-to-end customer churn prediction project, including business context, data preparation, exploratory analysis, modeling, and dashboard design. It is intended as a portfolio-ready case study for stakeholders, hiring managers, and technical reviewers.


## 1. Project overview

Telecommunications providers face significant revenue loss when customers cancel their service (churn). Acquiring new customers is typically more expensive than retaining existing ones, so understanding **who is likely to churn** and **why** is critical for designing effective retention strategies.

This project uses the Telco Customer Churn dataset to:

- Predict which customers are likely to churn.
- Identify the key drivers of churn behavior.
- Prepare a Power BI dashboard that surfaces churn risk and actionable insights for business stakeholders.

**Objective:**  
Build an interpretable, end-to-end churn prediction pipeline—from raw data to a production-ready dashboard dataset—that supports data-driven retention decisions.

## 2. Dataset summary

The Telco Customer Churn dataset contains information about customer demographics, account details, service usage, and billing. After cleaning, the dataset includes **7,032 customers** and **21 features**.

### **Key characteristics**
- **Target variable:** `Churn` (0 = stayed, 1 = churned)
- **Customer identifier:** `customerID`
- **Data sources:** Billing records, service subscriptions, demographic attributes

### **Feature groups**

#### **Numeric features**

- Partner  
- Dependents  
- tenure  
- PhoneService

- MonthlyCharges  
- TotalCharges  

#### **Categorical features**
- gender  
- MultipleLines  
- InternetService  
- OnlineSecurity  
- OnlineBackup  
- DeviceProtection  
- TechSupport  
- StreamingTV  
- StreamingMovies  
- Contract  
- PaymentMethod  

### **Engineered features**
To improve model performance, several engineered features were added:

- **tenure_group** — buckets customers into:  
  - 0–12 months  
  - 12–24 months  
  - 24–48 months  
  - 48–72 months  

- **One‑Hot Encoded categorical features** — e.g.:  
  - gender_Female, gender_Male  
  - Contract_Month-to-month, Contract_One year, Contract_Two year  
  - InternetService_DSL, InternetService_Fiber optic, InternetService_No  
  - PaymentMethod_* categories  

- **Scaled numeric features** using StandardScaler  
- **original_index** — preserves row alignment for merging predictions later

### **Final dataset structure**
After cleaning and engineering, the dataset is ready for:
- Exploratory Data Analysis (EDA)
- Machine learning modeling
- Dashboard integration

## 3. Key EDA insights

Exploratory Data Analysis (EDA) was performed to understand customer behavior, identify churn patterns, and validate assumptions before modeling. Several clear trends emerged from the data.

### **1. Churn rate**
The dataset shows a churn rate of **26.6%**, meaning roughly one in four customers leave the service. This confirms churn is a significant business problem.

### **2. Tenure is the strongest churn driver**
Customers with low tenure (0–12 months) churn at much higher rates.  
Long‑tenured customers are significantly more stable.

**Interpretation:**  
Early‑stage customers are at the highest risk and should be prioritized for onboarding and retention programs.

### **3. Contract type strongly influences churn**
- **Month‑to‑month contracts** have the highest churn.  
- **One‑year and two‑year contracts** show much lower churn.

**Interpretation:**  
Contract structure is a major retention lever. Encouraging longer-term contracts can reduce churn.

### **4. Payment method affects churn**
Customers using **Electronic Check** churn more frequently than those using automatic payment methods.

**Interpretation:**  
Electronic check users may experience billing friction or lack autopay convenience, increasing churn risk.

### **5. Internet service type matters**
- **Fiber optic customers** churn more than DSL customers.  
- Customers with **no internet service** churn the least.

**Interpretation:**  
Fiber optic customers may experience service issues or higher pricing, contributing to dissatisfaction.

### **6. Monthly charges correlate with churn**
Higher monthly charges are associated with higher churn rates.

**Interpretation:**  
Pricing pressure is a key churn driver. High‑cost customers may be more sensitive to service issues or billing changes.

### **7. Numeric correlations**
A correlation heatmap revealed:
- **tenure** has a strong negative correlation with churn  
- **MonthlyCharges** has a positive correlation with churn  
- Other numeric features show weak correlations

**Interpretation:**  
Churn is driven more by service and contract attributes than by demographic factors.

---

### **Summary of EDA findings**
- Low tenure → high churn  
- Month‑to‑month contracts → highest churn  
- Electronic check → highest churn  
- Fiber optic → higher churn than DSL  
- Higher monthly charges → higher churn  

These insights directly informed feature engineering, model selection, and dashboard design.


## 4. Modeling summary

To predict customer churn, three supervised machine learning models were trained using the engineered feature set:

- Logistic Regression (baseline)
- Random Forest Classifier
- XGBoost Classifier

All models were evaluated using:
- Accuracy
- Precision
- Recall
- F1-score
- ROC-AUC
- Confusion Matrix

### **Model performance comparison**

| Model                 | Accuracy | Recall (Churn) | ROC-AUC |
|-----------------------|----------|----------------|---------|
| Logistic Regression   | 0.80     | 0.57           | 0.836   |
| Random Forest         | 0.80     | 0.60           | 0.824   |
| XGBoost               | 0.78     | 0.53           | 0.820   |

### **Interpretation of results**

#### **Logistic Regression**
- Best overall ROC-AUC (0.836)
- Strong baseline performance
- Highly interpretable
- Stable across different thresholds

**Why it matters:**  
Logistic Regression provides a reliable probability score for churn, making it ideal for risk segmentation and dashboard use.

#### **Random Forest**
- Highest recall for churn (0.60)
- Strong performance on minority class
- Provides feature importance for interpretability

**Why it matters:**  
Random Forest helps identify the strongest churn drivers and supports business decision-making.

#### **XGBoost**
- Competitive performance
- Slightly lower recall and ROC-AUC
- Useful for comparison and validation

**Why it matters:**  
XGBoost confirms that tree-based models perform well but do not outperform simpler models on this dataset.

### **Selected model**
**Logistic Regression** was chosen as the primary model for churn probability scoring due to its:

- Best ROC-AUC performance  
- Smooth probability outputs  
- Interpretability  
- Stability across customer segments  

These probability scores were used to create churn risk segments for the dashboard.


## 5. Feature importance interpretation

Understanding *why* customers churn is just as important as predicting *who* will churn.  
Random Forest feature importance provides a clear, interpretable view of the strongest churn drivers.

### **Top churn drivers (Random Forest)**

1. **TotalCharges**  
   - Customers with lower lifetime spending churn more frequently.  
   - Indicates early‑stage customers are at higher risk.

2. **tenure**  
   - Short‑tenure customers churn at significantly higher rates.  
   - Tenure remains one of the most powerful predictors.

3. **MonthlyCharges**  
   - Higher monthly bills correlate with higher churn.  
   - Pricing pressure is a major churn factor.

4. **Contract_Month-to-month**  
   - Month‑to‑month customers churn far more than annual contract customers.  
   - Contract structure is a key retention lever.

5. **OnlineSecurity_No**  
   - Customers without online security services churn more.  
   - Suggests service bundle adoption reduces churn.

6. **TechSupport_No**  
   - Lack of technical support is associated with higher churn.  
   - Indicates support services play a role in customer satisfaction.

7. **InternetService_Fiber optic**  
   - Fiber optic customers churn more than DSL customers.  
   - May indicate service issues or higher pricing.

8. **PaymentMethod_Electronic check**  
   - Electronic check users churn at the highest rate.  
   - Billing friction or lack of autopay convenience may contribute.

### **Interpretation**

The model reveals that churn is driven by a combination of:

- **Customer lifecycle factors** (tenure, TotalCharges)  
- **Pricing factors** (MonthlyCharges)  
- **Contract structure** (month‑to‑month vs annual)

These insights directly inform retention strategies and dashboard design.

### **Business implications**

## 6. Dashboard dataset preparation

To support business decision-making, a Power BI dashboard was created to visualize churn risk, customer profiles, and key churn drivers. This required preparing a unified dataset that combines:

- Model predictions  
- Actual churn labels  
- Customer attributes  
- Engineered features  
- Risk segmentation  

### **1. Merging predictions with customer data**

The churn model produced probability scores for each customer in the test set. These predictions were merged with the cleaned dataset using `customerID`:

- Ensures correct alignment  
- Preserves customer attributes  
- Allows full customer profiles to be displayed in the dashboard  

The `original_index` column (created during feature engineering) guaranteed that predictions matched the correct customers.

### **2. Creating interpretable churn labels**

To make churn status readable in Power BI, a human-friendly label was added:

- `ChurnLabel = "Yes"` for churned customers  
- `ChurnLabel = "No"` for retained customers  

This improves clarity in visuals and filters.

### **3. Building churn risk segments**

Predicted probabilities were grouped into four risk levels:

| Probability Range | Segment     |
|-------------------|-------------|
| 0.00 – 0.25        | Low         |
| 0.25 – 0.50        | Medium      |
| 0.50 – 0.75        | High        |
| 0.75 – 1.00        | Critical    |

**Purpose:**  
Risk segmentation helps stakeholders quickly identify customers who need immediate retention action.

## 8. Business recommendations

Based on the exploratory analysis, model insights, and dashboard results, several clear opportunities emerge for reducing churn and improving customer retention. These recommendations are grounded in data and aligned with the strongest churn drivers identified in the project.

---

### **1. Strengthen early‑stage customer onboarding**
Customers with low tenure (0–12 months) churn at the highest rate.

**Recommendation:**  
Implement a structured onboarding program that includes:
- Welcome calls or emails  
- Setup assistance  
- Early check-ins  
- Personalized service recommendations  

**Expected impact:**  
Improves customer satisfaction during the critical early months and reduces churn risk.

---

### **2. Promote long-term contracts**
Month‑to‑month customers churn significantly more than annual contract customers.

**Recommendation:**  
Offer incentives for upgrading to one‑year or two‑year contracts:
- Discounts  
- Loyalty rewards  
- Bundled services  
- Contract upgrade promotions  

**Expected impact:**  
Stabilizes revenue and reduces churn by locking in longer commitments.

---

### **3. Encourage automatic payment methods**
Electronic check users churn at the highest rate.

Promote automatic payment options such as:
- Credit card autopay  
- Bank transfer autopay  

Offer small incentives for switching:
- Bill credits  
- Loyalty points  
- Fee waivers  

**Expected impact:**  
Reduces billing friction and improves retention.

Fiber optic customers churn more than DSL customers.

Improves satisfaction in a high-risk segment.
