### Telco Customer Churn — Data Cleaning Notebook

This notebook performs initial data quality checks, handles missing values,
fixes data types, removes duplicates, and prepares the dataset for EDA and
feature engineering.

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Initial shape and preview
df.shape
df.head()
df.info()
```

```python
# Check Missing Values
df.isna().sum()
```

```python
#Check Duplicate Customer IDs
df['customerID'].nunique(), df.shape[0]
```

```python
# Handling Whitespace
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'].isna().sum()
```

```python
# Removing NaN
df = df.dropna(subset=['TotalCharges'])
df.reset_index(drop=True, inplace=True)
df['TotalCharges'].isna().sum()
```

```python
#Converting Yes/No to Binary
binary_cols = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']

for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})
df.info()
```

```python
#Cleaning Senior Citizen Column
df['SeniorCitizen'] = df['SeniorCitizen'].astype(int)
df.info()
```

```python
# Identifying Categorical and Numeric Columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()

categorical_cols, numeric_cols
```

```python
df.to_csv("../data/cleaned/cleaned_telco_churn.csv", index=False)
```

# Summary

- Converted TotalCharges to numeric
- Removed rows with invalid TotalCharges
- Cleaned binary Yes/No columns
- Verified no duplicate customer IDs
- Identified categorical and numeric columns
- Exported cleaned dataset for EDA and feature engineering
